# MatplotLib
It is a ploting Library for python programming language and it's numerical mathematical extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI tools like Tkinter,wxPython,Qt or GTK+.

Advantages of Matplotlib are:
- Get started for simple plots
- Support for custom labels and texts
- Control of every element in a figure
- High-quality output in many formats
- very customizable in genral

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline 
```

```python
# No need to write plt.show()  to plot graph
```

```python
# Example
x = np.arange(0,10)
y = np.arange(11,21)
```

```python
## Plt scatter
plt.scatter(x,y,c='g')
```

```python
plt.scatter(x,y,c='g')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2D Graph')
# Save graph into jpg or png format
plt.savefig('test.png')
```

```python
y = x*x
```

```python
## plt plot
plt.plot(x,y)
```

```python
plt.plot(x,y,'r*--')
```

```python
plt.plot(x,y,'r*',linestyle='dashed',linewidth=2,markersize=12)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2D')
```

```python
### Subplots  e.g here 2x2=4 subplots
plt.subplot(2,2,1)  # 2-rows,2-column, 1st postion
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(x,y,'g*-')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'y--')
```

```python
x = np.arange(1,11)
y = 3 * x + 5
plt.title('Matplotlib demo')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.plot(x,y)
```

```python
# Sine plot
x = np.arange(0,4*np.pi,0.1)
y = np.sin(x)
plt.title('sine wave form')
plt.plot(x,y)
```

```python
## Subplot
# sine and cosine curves
x = np.arange(0,5*np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2,1,1)
plt.plot(x,y_sin,'r*')
plt.title('Sine Wave')

plt.subplot(2,1,2)
plt.plot(x,y_cos,'yo')
plt.title('Cosine Wave')
```

```python
# Bar plot
x = [2,8,10]
y = [11,16,9]

x2 = [3,9,11]
y2 = [6,15,7]


plt.bar(x,y,align='center')
plt.bar(x2,y2,color='g')

plt.title('Bar graph')
plt.ylabel('Y axis')
plt.xlabel('X axis')
```

### Histograms
- show density/count of the number
- bins=10 by default taken i.e like class intervel

```python
x = np.array([20,20,40,50,30,50,50,15,15,11,19,19,19,19,60,60,60,60,60,16])
plt.hist(x,color='r')
```

# Box Plot
- help us to find out percentile

```python
data = [np.random.normal(0,std,100) for std in range(1,4)]
data
```

```python
data = [np.random.normal(0,std,100) for std in range(1,4)]

plt.boxplot(data,vert=True,patch_artist=True)
```

# Pie Chart

```python
labels = 'Python','C++','Java','Blockchain'
sizes = [213,135,250,220]
colors = ['gold','yellowgreen','lightcoral','lightskyblue']
explode = (0.1,0,0,0) 

plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True)
plt.axis('equal')
```
