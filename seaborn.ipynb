# Seaborn
It is a visualization library used in Python. It is best use to know stastical techniques by visualizing data. And seaboarn are very useful while performing exploratary data analysis (EDA).
## Distribution Plot (Continuous plot)
- joinplot
- distplot
- pairplot

```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
# Our model have to constructed to find the tip based on other features
## tip: dependent feature
## total_bill,sex,smoker,day,time,size: independent features
```

## Correlation with Heatmap
Correlation helps us to find out inter-relationship between all the features.

Correlation will only be find out if our features are integer or floating values. Correlation will be not available for categorical features.

Correlation value will be ranges between [-1,+1] i.e called as pearson co-efficient co-relation

A correlation heatmap uses colored cells, typically in monochromatic scale, to show a 2D correlation matrix (table) between two discrete dimensions or event types. It is very important in Feature Selection.

```python
df.dtypes
```

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

## Joinplot
A join plot allows to study the relationship between two numeric variables. The centeral chart display their correlation. It is usually a scatterplot, a hexbin plot, a 2D histogram or a 2D density plot.


### Bivariate analysis

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

## Pair plot
If we have more than two independent features the at that time we use pair plot.

A 'pair plot' is also known as a scatterplot, in which one variable in the same data row is matched with another variable's value value, like this: Pairs plots are just elaborations on this, showing all variables paired with all the other variables.

```python
sns.pairplot(df)
```

```python
# We can also plot pair plot based on some feature like sex,day,time etc
sns.pairplot(df,hue='sex')
```

```python
sns.pairplot(df,hue='smoker')
```

```python
df['smoker'].value_counts()
```

## Dist plot
Dist plot help us to check the distribution of the columns feature.

It basically create histogram of the feature.

```python
sns.displot(df['tip'],kde=True)  #kde=kernal density estimate
```

```python
sns.displot(df['tip'],kde=False,bins=10)
```

## Categorical Plots
Seaborn also helps us in doing the analysis on categorical data points.
- boxplot
- violinplot
- countplot
- bar plot

```python
sns.countplot('day',data=df)
```

```python
sns.countplot(y='day',data=df)
```

```python
# Bar plot
sns.barplot(x='sex',y='total_bill',data=df)
```

## Box plot
A box and whisker plot (sometimes called a boxplot) is a graph that presents information form a five-number summary

```python
sns.boxplot('sex','total_bill',data=df)
```

```python
sns.boxplot(x='day',y='total_bill',data=df,palette='rainbow')
```

```python
sns.boxplot(data=df,orient='v')
```

```python
# categorize boxplot based on some other category
sns.boxplot(x='total_bill',y='day',hue='smoker',data=df)
```

## Violin Plot
Violin plot helps us to see both the distribution of data in terms of kernal density estimation and the box plot

```python
sns.violinplot(x='total_bill',y='day',data=df,palette='rainbow')
```
